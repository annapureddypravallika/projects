```python
import numpy as  np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as ans;
from google.colab import drive
drive.mount('/content/drive')
df=pd.read_csv("/content/drive/MyDrive/heart.csv")
df.head()
```

```python
#show correlation matrix
corr=df.corr()
plt.figure(figsize=(18,10))
ans.heatmap(corr,annot=True,)
plt.show()
```

```python
#plot target dist.
#ans.countplot(df.target,hue='sex',palette=['green','red'])
#ans.countplot(x=df.target,hue='sex',palette=['#1CA53B','red'])
ans.countplot(x=df.target,hue='sex',data=df,palette=['#1CA53B','red'])
#seaborn.pairplot(df,hue='size')
plt.xticks([0,1],labels=['not Disease','Disease'])
plt.title("Target Distribution");
```

```python
#plot target dist.
#ans.countplot(df.target,hue='sex',palette=['green','red'])
#ans.countplot(x=df.target,hue='sex',palette=['#1CA53B','red'])
ans.countplot(x=df.target,data=df,palette=['#1CA53B','red'])
#seaborn.pairplot(df,hue='size')
plt.xticks([0,1],labels=['not Disease','Disease'])
plt.title("Target Distribution");
```

```python
plt.figure(figsize=(18,10))
ans.countplot(x='age',hue='target',data=df,palette=['#1CA53B','red'])
plt.legend("Haven't Disease","Have Disease")
plt.title("Heart Disease Frequency for Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(18,10))
ans.countplot(x='age',hue='target',data=df)
plt.legend("Haven't Disease","Have Disease")
plt.title("Heart Disease Frequency for Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#ans.set_style("whitegrid")
plt.figure(figsize=(18,10))
ans.displot(df.age[df['target']==0],bins=30,color='#1CA53B',label='Haven\'t Disease')
ans.displot(df.age[df['target']==1],bins=30,color='red',label='Have Disease')
plt.legend()
plt.title("Heart Disease Frequency for Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#ans.set_style("whitegrid")
plt.figure(figsize=(18,10))
ans.distplot(df.age[df['target']==0],bins=30,color='Black')
ans.distplot(df.age[df['target']==1],bins=30,color='pink')
plt.legend()
plt.title("Heart disease distributin for ages")
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.show()
```
